Create parallel sentences dataframe in a single file for match and almost match and also a file contaning statistical data

```python
rerun=False
if not rerun:
    start_from=0
    ap_sen_csv=pd.DataFrame(columns=['file_no','m_or_am','english_sen','hindi_sen'])
    count=0
    pd_sen_data=pd.DataFrame(columns=['file_no','match_total_en','match_total_hi','match_en','match_hi',\
                                      'almost_match_en','almost_match_hi'])
else:
    start_from=0 # if rerunning set value from where the run failed
    
fl_list=glob.glob(align_loc+'//'+'*.txt')
m_fl_en=sorted([os.path.basename(i) for i in fl_list if len([k for k in ['m','en'] if \
                                                             k in os.path.basename(i).split('_')])==2])
am_fl_en=sorted([os.path.basename(i) for i in fl_list if len([k for k in ['am','en'] if \
                                                              k in os.path.basename(i).split('_')])==2])
c_fl_en=list(zip(m_fl_en,am_fl_en))
m_fl_hi=sorted([os.path.basename(i) for i in fl_list if len([k for k in ['m','hi'] if \
                                                             k in os.path.basename(i).split('_')])==2])
am_fl_hi=sorted([os.path.basename(i) for i in fl_list if len([k for k in ['am','hi'] if \
                                                              k in os.path.basename(i).split('_')])==2])
c_fl_hi=list(zip(m_fl_hi,am_fl_hi))
c_fl=list(zip(c_fl_en,c_fl_hi))

for k,fl_name in enumerate(c_fl[start_from:]):
    i=k+start_from
    pd_sen_data.loc[i]=np.nan
    ap_sen_csv.loc[count]=np.nan
    pd_sen_data.at[i,'file_no']= fl_name[0][0].split('_')[2]
    #open english match file
    if (get_file_content(os.path.join(align_loc,fl_name[0][0])) and get_file_content(os.path.join(align_loc,\
                                                                                                 fl_name[1][0]))):
        data_m_en=get_file_content(os.path.join(align_loc,fl_name[0][0]))
        pd_sen_data.at[i,'match_en']=data_m_en.count('\n')
    #open corresponding hindi match file
        data_m_hi=get_file_content(os.path.join(align_loc,fl_name[1][0]))
        pd_sen_data.at[i,'match_hi']=data_m_hi.count('\n')
    #add sentences
        data_match=list(zip(data_m_en.split('\n')[:-1],data_m_hi.split('\n')[:-1]))
        for k in data_match:
            ap_sen_csv.loc[count,:]=[fl_name[0][0].split('_')[2],'m',k[0],k[1]]
            count+=1
    #print(data_match)
    #open english almost match file
    if (get_file_content(os.path.join(align_loc,fl_name[0][1])) and get_file_content(os.path.join(align_loc,\
                                                                                                  fl_name[1][1]))):
        data_am_en=get_file_content(os.path.join(align_loc,fl_name[0][1]))
        pd_sen_data.at[i,'almost_match_en']=data_am_en.count('\n')
        #open hindi almost match file
        data_am_hi=get_file_content(os.path.join(align_loc,fl_name[1][1]))
        pd_sen_data.at[i,'almost_match_hi']=data_am_hi.count('\n')
        #add sentences
        data_almost_match=list(zip(data_am_en.split('\n')[:-1],data_am_hi.split('\n')[:-1]))
        for k2 in data_almost_match:
            ap_sen_csv.loc[count,:]=[fl_name[0][0].split('_')[2],'am',k2[0],k2[1]]
            count+=1
    #add totals
pd_sen_data.fillna(0,inplace=True)
pd_sen_data.loc[:,'match_total_hi']=pd_sen_data.loc[:,'match_hi']+pd_sen_data.loc[:,'almost_match_hi']
pd_sen_data.loc[:,'match_total_en']=pd_sen_data.loc[:,'match_en']+pd_sen_data.loc[:,'almost_match_en']
```

```python
#Working on paralle sentences datafram-----------final data stored in ap_sen_csv
    # Drop Duplicate sentences
ap_sen_csv.drop_duplicates(subset=['english_sen'],inplace=True)
    #Remove sentences with non english content in english line
if 'contain_ot_eng' not in ap_sen_csv.columns.tolist():
    ap_sen_csv['contain_ot_eng']=ap_sen_csv['english_sen'].apply(lambda x: detect_non_eng(x))
ap_sen_csv.drop(ap_sen_csv[ap_sen_csv['contain_ot_eng']==True].index,inplace=True)
    #Reset Index
ap_sen_csv.reset_index(drop=True,inplace=True)
                         
#Working on stats file-------final data is stored in pd_new
    #checking for any inconsistency in data
if 'match_total_hi'  in pd_sen_data.columns.tolist(): 
    if (pd_sen_data.query('match_en!=match_hi').shape[0]==0 and \
        pd_sen_data.query('almost_match_en!=almost_match_hi').shape[0]==0):

        pd_sen_data.loc[:,"match_total"]=pd_sen_data.loc[:,"match_total_en"]
        pd_sen_data.loc[:,"match"]=pd_sen_data.loc[:,"match_en"]
        pd_sen_data.loc[:,"almost_match"]=pd_sen_data.loc[:,"almost_match_en"]
        pd_sen_data.drop(["match_total_hi","match_total_en","match_hi","match_en","almost_match_en","almost_match_hi"],\
                     axis=1,inplace=True)
    else: print('Check null values')

pd_data_tok=pd.read_csv(work_dir+'//'+'tok_sen_count_'+'_'.join([n_month,n_year])+'.csv',encoding='utf-16')
pd_data_tok.loc[:,'file_no']=pd_data_tok.loc[:,'Filename_en'].apply(lambda x:str(x).split('_')[0])
pd_new=pd.merge(pd_sen_data,pd_data_tok,left_on='file_no',right_on='file_no',how='left')
```
